# Unsupervised Learning (Clustering)

## Step 1: Prep Data

```python
# Create a field within our dataset named text that we will use for the embeddings
df = names
df = df.dropna(subset = 'categories') # This shouldn't do anything, since we know there are no NaN from earlier
df['text'] = df['categories'].astype(str)
```

```python
from sentence_transformers import SentenceTransformer

model = SentenceTransformer('all-MiniLM-L6-v2')
embeddings = model.encode(df['text'].tolist(), show_progress_bar = True)
```

## Step 3: Reduce Dimensions

Let's reduce dimensions using a k-means cluster. We do not necessarily need to do a cluster, but we will perform it now to see how this initial step in partitioning the data works.

## Step 6: Clean Up the Data and Rerun the Model

The code above gives an idea on how to approach the clustering, but it's simple. A silhouette score of 0.36 isn't inherently bad, but it's not particularly good either. Let's pre-process the category column and run it through the same model again to see if this will work

```python
import re # Provides regular expression (regex) support
from sklearn.feature_extraction.text import ENGLISH_STOP_WORDS # NLTK has a more comprehensive list

# Define stopwords by building on the scikit learn list
base_stopwords = set(ENGLISH_STOP_WORDS)

stopwords = base_stopwords.union({
    'baby', 'product', 'products', 'infant', 'newborn', 'brand', 'pack', 'size',
    'new'
})

# Function to clean up stopwords
def clean_text(text):
    if not isinstance(text, str):
        return ""

    # Remove square brackets, squiggly brackets, colon, semi-colon, and comma
    text = re.sub(r'[\[\]\{\}\:\;\,]', '', text)

    # Remove quotes
    text = text.replace("'", "").replace('"', '')

    # Normalize whitespace
    text = re.sub(r'\s+', ' ', text)

    # Lowercase
    text = text.lower()

    # Split and remove stopwords
    words = text.split()
    filtered_words = [word for word in words if word not in stopwords]

    return ' '.join(filtered_words).strip()
```

```python
# Apply the cleanup to the fields I am using for clustering

# Redefine df 
df = names

# Create a categories column that is a string. 
# Fill missing values first and convert all values to lowercase
df['categories_clean'] = df['categories'].astype('string').fillna('').str.lower()

# Optional: Truncate very long text fields
df['categories_clean'] = df['categories_clean'].str.slice(0, 200)

# Apply the defined function in the previous chunk to the categories_clean column
df['categories_clean'] = df['categories_clean'].apply(clean_text)

# Confirm stopwords were removed
# df[df['title'].isin(stopwords)]
check = df['categories_clean'].value_counts(dropna=False).reset_index()
check.columns = ['categories_clean', 'count']
check['percentage'] = (category_freq['count'] / category_freq['count'].sum()) * 100

check.head(100)
```

# Step 7: Refine Cluster Creation

After working with both the raw data and attempting to clean it up using stopwords, the model itself via the silhouette score in conjunction with research on models like BERT indicates that the stop words should remain.

The presence of special characters is less clear. For now, let's remove special characters and spaces, as well as clean up case sensitivity. If the model still seems lacking with special characters removed, they can be added in. Additionally, some meaningless words in this case, such as baby and products, can be removed since they do not add value to the descriptions.

Additionally, the process of cleaning '[]' categories should continue as thse do not produce meaningful results (how can we accurately interpret [] as any sort of baby product).

```python
# Prepare the categories column for KMeans clustering. Specifically, if it contains meaningless text, such as [], 
# replace this with more meaningful text 

# Redefine df 
df = names

# Create a function to replace missing or '[]' values in the categories_clean column
def cat_replacement(row):
    for col in ['categories', 'title', 'description', 'details']:
        val = str(row[col]).strip()
        if val not in ['', '[]']:
            return val
    return '' # fallback if all are blank

# Call the function
df['category_select'] = df.apply(cat_replacement, axis = 1)

# Make sure the categories column is a string. Fill missing values first and convert all values to lowercase
df['categories_clean'] = df['category_select'].astype('string').fillna('').str.lower()

# Print check
df.head(100)
```

```python
# Confirm that none of the category values are blank or []
df[df['categories_clean'].isin(["", "[]"])]
```

```python
# Define process for removing special characters from the dataset
import re # Provides regular expression (regex) support

# Define domain specific stopwords: 

stopwords = {
    'baby', 'product', 'products', 'infant', 'newborn', 'brand', 'pack', 'size',
    'new'
}

# Function to clean special characters and domain specific stopwords
def clean_text(text):
    if not isinstance(text, str):
        return ""

    # Remove square brackets, squiggly brackets, colon, semi-colon, and comma
    text = re.sub(r'[\[\]\{\}\:\;\,]', '', text)

    # Remove quotes
    text = text.replace("'", "").replace('"', '')

    # Normalize whitespace
    text = re.sub(r'\s+', ' ', text)

    # Lowercase
    text = text.lower()

    # Split words (may not be necessary anymore now that we're not using stopwords )
    words = text.split()
    filtered_words = [word for word in words if word not in stopwords]

    return ' '.join(filtered_words).strip()
```

```python
# Remove special characters and whitespace from the categories_clean field
df['categories_clean'] = df['categories_clean'].apply(clean_text)
df.head(100)
```

```python
# Confirm that none of the results are empty of []
df[df['categories_clean'].isin(["", "[]"])]
```

```python
# Now we can get back to applying the model

# Create a field within our dataset named text that we will use for the embeddings
df = df.dropna(subset = 'categories_clean') # This shouldn't do anything, since we know there are no NaN from earlier
df['text'] = df['categories_clean'].astype(str)

# Run BERT for the embeddings/vector representations
from sentence_transformers import SentenceTransformer

#model = SentenceTransformer('all-MiniLM-L6-v2')
model = SentenceTransformer('all-MiniLM-L6-v2')
embeddings = model.encode(df['text'].tolist(), show_progress_bar = True)
```

```python
# Conduct unsuperviced learning on the embedded data
# Reduce dimensions
from sklearn.decomposition import PCA

# Reduce to 50 dimensions for better KMeans performance
pca = PCA(n_components = 50)
reduced_embeddings = pca.fit_transform(embeddings)

# Cluster using KMeans
from sklearn.cluster import KMeans

# Use more clusters
kmeans = KMeans(n_clusters = 20, random_state = 42, n_init = 10) # n_init is the number of times KMeans will run with different centroid seeds
df['cluster'] = kmeans.fit_predict(reduced_embeddings)
```

# Step 8: Utilize UMAP Instead of PCA for Embedding

Some additional research indicates that UMAP may work better than PCA necause BERT is nonlinear and UMAP works better with non-linear data while PCA works more with linear data.

```python
# Prepare data for KMeans clustering using UMAP for dimension reductionality
# Redefine df 
df = names

# Create a function to replace missing or '[]' values in the categories_clean column
def cat_replacement(row):
    for col in ['categories', 'title', 'description', 'details']:
        val = str(row[col]).strip()
        if val not in ['', '[]']:
            return val
    return '' # fallback if all are blank

# Call the function
df['category_select'] = df.apply(cat_replacement, axis = 1)

# Make sure the categories column is a string. Fill missing values first and convert all values to lowercase
df['categories_clean'] = df['category_select'].astype('string').fillna('').str.lower()

# Define process for removing special characters from the dataset
import re # Provides regular expression (regex) support

# Define domain specific stopwords: 

stopwords = {
    'baby', 'product', 'products', 'infant', 'newborn', 'brand', 'pack', 'size',
    'new'
}

# Function to clean special characters and domain specific stopwords
def clean_text(text):
    if not isinstance(text, str):
        return ""

    # Remove square brackets, squiggly brackets, colon, semi-colon, and comma
    text = re.sub(r'[\[\]\{\}\:\;\,]', '', text)

    # Remove quotes
    text = text.replace("'", "").replace('"', '')

    # Normalize whitespace
    text = re.sub(r'\s+', ' ', text)

    # Lowercase
    text = text.lower()

    # Split words (may not be necessary anymore now that we're not using stopwords )
    words = text.split()
    filtered_words = [word for word in words if word not in stopwords]

    return ' '.join(filtered_words).strip()

# Remove special characters and whitespace from the categories_clean field
df['categories_clean'] = df['categories_clean'].apply(clean_text)
```

```python
# Apply BERT word embeddings
# Create a field within our dataset named text that we will use for the embeddings
df = df.dropna(subset = 'categories_clean') # This shouldn't do anything, since we know there are no NaN from earlier
df['text'] = df['categories_clean'].astype(str)

# Run BERT for the embeddings/vector representations
from sentence_transformers import SentenceTransformer

model = SentenceTransformer('all-MiniLM-L6-v2')
embeddings = model.encode(df['text'].tolist(), show_progress_bar = True)
```

```python
# Dimensionality reduction using UMAP - key difference to the previous steps
from umap import UMAP

umap_model = UMAP(n_components=15, random_state=42)  # Try 2 for plotting, 15‚Äì50 for clustering
reduced_embeddings = umap_model.fit_transform(embeddings)

# Cluster using KMeans
from sklearn.cluster import KMeans

# Perform the actual clustering
kmeans = KMeans(n_clusters = 20, random_state = 42, n_init = 10) # n_init is the number of times KMeans will run with different centroid seeds
df['cluster'] = kmeans.fit_predict(reduced_embeddings)
```

```python
# Evaluate the model to see gauge its performance with these new updates
from sklearn.metrics import silhouette_score

score = silhouette_score(reduced_embeddings, df['cluster'])
print(f"Silhouette Score: {score:.2f}")
```

```python
# Prepare data for KMeans clustering using UMAP for dimension reductionality
# Redefine df 
df = names

# Create a function to replace missing or '[]' values in the categories_clean column
def cat_replacement(row):
    for col in ['categories', 'title', 'description', 'details']:
        val = str(row[col]).strip()
        if val not in ['', '[]']:
            return val
    return '' # fallback if all are blank

# Call the function
df['category_select'] = df.apply(cat_replacement, axis = 1)

# Make sure the categories column is a string. Fill missing values first and convert all values to lowercase
df['categories_clean'] = df['category_select'].astype('string').fillna('').str.lower()

# Define process for removing special characters from the dataset
import re # Provides regular expression (regex) support

# Define domain specific stopwords: 

stopwords = {
    'baby', 'product', 'products', 'infant', 'newborn', 'brand', 'pack', 'size',
    'new'
}

# Function to clean special characters and domain specific stopwords
def clean_text(text):
    if not isinstance(text, str):
        return ""

    # Remove square brackets, squiggly brackets, colon, semi-colon, and comma
    text = re.sub(r'[\[\]\{\}\:\;\,]', '', text)

    # Remove quotes
    text = text.replace("'", "").replace('"', '')

    # Normalize whitespace
    text = re.sub(r'\s+', ' ', text)

    # Lowercase
    text = text.lower()

    # Split words (may not be necessary anymore now that we're not using stopwords )
    words = text.split()
    filtered_words = [word for word in words if word not in stopwords]

    return ' '.join(filtered_words).strip()

# Remove special characters and whitespace from the categories_clean field
df['categories_clean'] = df['categories_clean'].apply(clean_text)
```
