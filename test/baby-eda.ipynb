# Unsupervised Learning (Clustering)

## Step 1: Prep Data

```python
# Create a field within our dataset named text that we will use for the embeddings
df = names
df = df.dropna(subset = 'categories') # This shouldn't do anything, since we know there are no NaN from earlier
df['text'] = df['categories'].astype(str)
```

## Step 2: Generate Embeddings using BERT

```python
from sentence_transformers import SentenceTransformer

model = SentenceTransformer('all-MiniLM-L6-v2')
embeddings = model.encode(df['text'].tolist(), show_progress_bar = True)
```

## Step 3: Reduce Dimensions

Let's reduce dimensions using a k-means cluster. We do not necessarily need to do a cluster, but we will perform it now to see how this initial step in partitioning the data works.

```python
from sklearn.decomposition import PCA

# Reduce to 50 dimensions for better KMeans performance
pca = PCA(n_components = 50)
reduced_embeddings = pca.fit_transform(embeddings)
```

## Step 4: Clustering Using KMeans

```python
from sklearn.cluster import KMeans

# Start with 10 clusters
kmeans = KMeans(n_clusters = 10, random_state = 42, n_init = 10) # n_init is the number of times KMeans will run with different centroid seeds
df['cluster'] = kmeans.fit_predict(reduced_embeddings)
```

## Step 5: Evaluate Cluster Quality

```python
from sklearn.metrics import silhouette_score

score = silhouette_score(reduced_embeddings, df['cluster'])
print(f"Silhouette Score: {score:.2f}")
```

## Step 6: Clean Up the Data and Rerun the Model

The code above gives an idea on how to approach the clustering, but it's simple. A silhouette score of 0.36 isn't inherently bad, but it's not particularly good either. Let's pre-process the category column and run it through the same model again to see if this will work

```python
import re # Provides regular expression (regex) support
from sklearn.feature_extraction.text import ENGLISH_STOP_WORDS # NLTK has a more comprehensive list

# Define stopwords by building on the scikit learn list
base_stopwords = set(ENGLISH_STOP_WORDS)

stopwords = base_stopwords.union({
    'baby', 'product', 'products', 'infant', 'newborn', 'brand', 'pack', 'size',
    'new'
})

# Function to clean up stopwords
def clean_text(text):
    if not isinstance(text, str): # Deals with floats, integers, NaN, or None
        return ""

    # Remove square brackets and their contents
    text = re.sub(r'\[.*?\]', '', text)

    # Remove newline characters and extra whitespace
    text = text.replace('\n', ' ').replace('\r', ' ')
    text = re.sub(r'\s+', ' ', text)

    # Lowercase everything
    text = text.lower()

    # Remove common stopwords (custom + general)
    words = text.split()
    filtered_words = [word for word in words if word not in stopwords]
    text = ' '.join(filtered_words)

    return text.strip()
```

```python
# Apply the cleanup to the fields I am using for clustering

# Redefine df 
df = names

# Fill missing values first
df['categories_clean'] = df['categories'].fillna('')

# Optional: Truncate very long text fields
df['categories_clean'] = df['categories_clean'].str.slice(0, 200)
```

```python
# Rerun all of the previous steps on this newly processed column

# Create a field within our dataset named text that we will use for the embeddings
df = df.dropna(subset = 'categories_clean') # This shouldn't do anything, since we know there are no NaN from earlier
df['text'] = df['categories_clean'].astype(str)

# Run BERT model
from sentence_transformers import SentenceTransformer

model = SentenceTransformer('all-MiniLM-L6-v2')
embeddings = model.encode(df['text'].tolist(), show_progress_bar = True)

# Reduce dimensions
from sklearn.decomposition import PCA

# Reduce to 50 dimensions for better KMeans performance
pca = PCA(n_components = 50)
reduced_embeddings = pca.fit_transform(embeddings)

# Cluster using KMeans
from sklearn.cluster import KMeans

# Start with 10 clusters
kmeans = KMeans(n_clusters = 10, random_state = 42, n_init = 10) # n_init is the number of times KMeans will run with different centroid seeds
df['cluster'] = kmeans.fit_predict(reduced_embeddings)
```

```python
# Evaluate the model to see if it improved
from sklearn.metrics import silhouette_score

score = silhouette_score(reduced_embeddings, df['cluster'])
print(f"Silhouette Score: {score:.2f}")
```

```python
# We get the same silhouette score as before. Let's run a for look that will iterate over several cluster sizes 
# and see if this fixes anything
for k in range(5, 21):  # Try a range of values
    km = KMeans(n_clusters=k, random_state=42, n_init=10)
    labels = km.fit_predict(reduced_embeddings)  # or raw embeddings
    score = silhouette_score(reduced_embeddings, labels)
    print(f"k={k} ‚Üí Silhouette Score: {score:.3f}")
```

```python
# Look at sample items from each cluster to name them
for c in sorted(df['cluster'].unique()):
    print(f"\nCluster {c}")
    print(df[df['cluster'] == c]['text'].sample(3, random_state=1).to_string(index=False))

# Visualize the clusters from seaborn
from sklearn.decomposition import PCA
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Reduce to 2D for visualization
pca_2d = PCA(n_components=2)
reduced_2d = pca_2d.fit_transform(embeddings)

# Create a DataFrame for plotting
plot_df = pd.DataFrame({
    'PC1': reduced_2d[:, 0],
    'PC2': reduced_2d[:, 1],
    'cluster': df['cluster'].astype(str)  # for coloring
})

# Plot with Seaborn
plt.figure(figsize=(10, 6))
sns.scatterplot(data=plot_df, x='PC1', y='PC2', hue='cluster', palette='tab10', alpha=0.7)
plt.title("2D PCA Visualization of Clusters")
plt.legend(title='Cluster')
plt.tight_layout()
plt.show()
```
